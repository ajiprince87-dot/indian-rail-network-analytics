#**Dataset Information**

**Source:** National Open Data Portal of India (`data.gov.in`)  
**Link:** [https://www.data.gov.in/](https://www.data.gov.in/)

**Year / Timeline:** Data updated and released via open government protocols representing active network parameters.

**Dataset Column/features Description:**
* **Train No** - Unique identification numeric code for each train.
* **Train Name** - Descriptive name of the specific rail asset.
* **SEQ** - Sequential halt sequence order number for the stations on the route.
* **Station Code** - Unique short alpha-code for each tracking station.
* **Station Name** - Complete localized name of the station.
* **Arrival time / Departure Time** - Scheduled operational timing grid matrices.
* **Distance** - Cumulative distance traveled in Kilometers from the origin station.
* **Source Station / Source Station Name** - Starting location attributes of the train.
* **Destination Station / Destination Station Name** - Ending terminal attributes of the train.
